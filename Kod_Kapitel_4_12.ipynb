```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
# clf som jag tagit fram från dokumentationen är förkortning för classifier.

clf = tree.DecisionTreeClassifier(random_state=42)
iris = load_iris()
iris
```

```python
# Träna modellen på iris datan.
# iris.data är blommans mått (input features) och iris.target är blommans typer (korrekt labeler).
clf = clf.fit(iris.data, iris.target)

# Jag ville skapa en större bild av trädet.
plt.figure(figsize=(13,9))
# Det visar namnen på input features och klasserna, samt färgkodar bladen i trädet.
tree.plot_tree(clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```
